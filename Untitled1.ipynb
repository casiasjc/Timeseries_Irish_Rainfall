```python
import pandas as pd
import numpy as np
from neuralprophet import NeuralProphet
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
df = pd.read_csv('Data/hrly_Irish_weather.csv')
df.head()
```

```python
df.station.unique()
```

```python
# Set stations
df = df[df['station'].isin(['VALENTIA OBSERVATORY', 'DUBLIN AIRPORT', 'GURTEEN'])]

# Set columns
df = df.loc[:, ['station', 'date', 'rain', 'temp', 'wdsp']]

# Set to datetime
df['date'] = pd.to_datetime(df['date'])
df.head()
```

```python
# Replace empty strings with NaN in all columns
df = df.replace(' ', np.nan)
```

```python
# Count NaN values in all columns by year
nan_counts_by_year = df.groupby(lambda x: df.loc[x, 'date'].year).apply(lambda x: x.isna().sum())

print(nan_counts_by_year)
```

```python
# create date range for years 2016-2020
start_date = '2016-01-01'
end_date = '2020-12-31'
date_range = pd.date_range(start=start_date, end=end_date, freq='D')

# Filter dataframe using boolean indexing
df = df[df['date'].isin(date_range)]
```

```python
# Convert columns to numerical
df['rain'] = pd.to_numeric(df['rain'], errors='coerce')
df['temp'] = pd.to_numeric(df['temp'], errors='coerce')
df['wdsp'] = pd.to_numeric(df['wdsp'], errors='coerce')
df.head()
```

```python
# Sanity check
df.dtypes
```

```python
# Fill in wdsp missing values with mode
mode_value = df['wdsp'].mode()[0]
df['wdsp'] = df['wdsp'].fillna(mode_value)

# Fill in rain missing values with interpolation technique
df['rain'] = df['rain'].interpolate(method='linear') 

# Fill in temp missing values with interpolation technique
df['temp'] = df['temp'].interpolate(method='linear')
```

```python
# Sanity check! Count NaN values in all columns
nan_counts = df.isna().sum()

print(nan_counts)
```

```python
# New dataframes by station
val_obs = df[df['station'] == 'VALENTIA OBSERVATORY']
dub_air = df[df['station'] == 'DUBLIN AIRPORT']
gurteen = df[df['station'] == 'GURTEEN']

# New dataframes by rain column 
val_obs_rain = val_obs[['date', 'rain']].copy()
dub_air_rain = dub_air[['date', 'rain']].copy()
gurteen_rain = gurteen[['date', 'rain']].copy()

# New dataframes by temp column 
val_obs_temp = val_obs[['date', 'temp']].copy()
dub_air_temp = dub_air[['date', 'temp']].copy()
gurteen_temp = gurteen[['date', 'temp']].copy()

# New dataframes by wdsp column 
val_obs_wdsp = val_obs[['date', 'wdsp']].copy()
dub_air_wdsp = dub_air[['date', 'wdsp']].copy()
gurteen_wdsp = gurteen[['date', 'wdsp']].copy()
```

## Valencia Observatory: Rain

```python
# Find the index which allows us to split off 20% of the data
cutoff_np_vor = round(np_vor.shape[0]*0.8)
cutoff_np_vor
```

```python
# Define train and test sets according to the index found above
train_np_vor = np_vor[:cutoff_np_vor]
test_np_vor = np_vor[cutoff_np_vor:]

# Print lengths
print("Train length:", len(train_np_vor))
print("Test length:", len(test_np_vor))
```

```python
# Plot of our train test split
fig, ax = plt.subplots(figsize=(20, 8))

ax.plot(train_np_vor.ds, label='Train')
ax.plot(test_np_vor.ds, label='Test')
ax.set_title('Train-Test Split');
plt.legend();
```

```python
# Initialize and fit the model on the training data
m_np_vor = NeuralProphet()

m_np_vor.fit(train_np_vor, freq='H', epochs=20)
```

```python
future_np_vor = m_np_vor.make_future_dataframe(np_vor, periods=900)
forecast_np_vor = m_np_vor.predict(future_np_vor)
```

```python
# Turn Test data back into a datetime dataframe so we can plot it easily
test_np_vor.set_index('ds', inplace=True)
test_np_vor.head()
```

```python
# Extract actual values from the test set
y_true = test_np_vor['y']

# Extract predicted values from the forecast
y_pred = forecast_np_vor['yhat1'].iloc[:len(y_true)]

# Compute MAE and RMSE
mae = mean_absolute_error(y_true, y_pred)
rmse = mean_squared_error(y_true, y_pred, squared=False)

# Print results
print("MAE: {:.4f}".format(mae))
print("RMSE: {:.4f}".format(rmse))
```
