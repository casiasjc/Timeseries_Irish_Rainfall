## Baseline Understanding: Naive Model

#### Rain

```python
# find the index which allows us to split off 20% of the data
cutoff_rain = round(dub_air_rain.shape[0]*0.8)

# Define train and test sets according to the index found above
train_rain = dub_air_rain[:cutoff_rain]
test_rain = dub_air_rain[cutoff_rain:]

# Print lengths
print("Train length:", len(train_rain))
print("Test length:", len(test_rain))
```

```python
# Calculate mean of rainfall in the training set
mean_rainfall = np.mean(train_rain["rain"])

# Create naive model that predicts mean rainfall for every input in the test set
naive_predictions = np.full_like(test_rain["rain"], mean_rainfall)

# Calculate and print mean squared error of naive model
mse_naive_rain = np.mean((test_rain["rain"] - naive_predictions)**2)
rmse_naive_rain = np.sqrt(mse_naive_rain)

print("Naive model MSE:", mse_naive_rain)
print("Naive model RMSE:", rmse_naive_rain)
```

#### Temp

```python
# find the index which allows us to split off 20% of the data
cutoff_temp = round(dub_air_temp.shape[0]*0.8)

# Define train and test sets according to the index found above
train_temp = dub_air_temp[:cutoff_temp]
test_temp = dub_air_temp[cutoff_temp:]

# Print lengths
print("Train length:", len(train_temp))
print("Test length:", len(test_temp))
```

```python
# Calculate mean of rainfall in the training set
mean_temperature = np.mean(train_temp['temp'])

# Create naive model that predicts mean rainfall for every input in the test set
naive_temp_predictions = np.full_like(test_temp['temp'], mean_temperature)

# Calculate and print mean squared error of naive model
mse_naive_temp = np.mean((test_temp['temp'] - naive_temp_predictions)**2)
rmse_naive_temp = np.sqrt(mse_naive_temp)

print("Naive model MSE:", mse_naive_temp)
print("Naive model RMSE:", rmse_naive_temp)
```

# First Model - Random Walk Model

#### Rain

```python
# New dataframe for train-test split
df_rain = dub_air_rain.set_index("date")

# find the index which allows us to split off 20% of the data
cutoff_rwr = round(df_rain.shape[0]*0.8)

# Define train and test sets according to the index found above
train_rwr = df_rain[:cutoff_rwr]
test_rwr = df_rain[cutoff_rwr:]

# Print lengths
print("Train length:", len(train_rwr))
print("Test length:", len(test_rwr))
```

```python
# Convert dataframe to numpy array with float data type
train_rwr_array = train_rwr.to_numpy(dtype=np.float)

# Fit an ARIMA(0,1,0) model
RW_rain = ARIMA(train_rwr_array, order=(0,1,0)).fit()

# Print out summary information on the fit
print(RW_rain.summary())
```

```python
#Root Mean Square Error(RMSE) For our Random Walk model
y_hat_rwr = RW_rain.predict(typ='levels')

RW_rain_rmse = mean_squared_error(train_rwr, y_hat_rwr, squared=False)

print(RW_rain_rmse)
```

```python
# Generate predictions on the test set
test_rwr_array = test_rwr.to_numpy(dtype=np.float)
y_hat_test_rwr = RW_rain.forecast(steps=len(test_rwr_array))

# Calculate RMSE on the test set
test_rwr_rmse = mean_squared_error(test_rwr_array, y_hat_test_rwr, squared=False)

print("Test RMSE:", test_rwr_rmse)
```

## Fifth Model - Neural Prophet Model

#### Dataframes

```python
# Dublin Airport Rain
np_dar = dub_air_rain[['date', 'rain']]
np_dar.columns = ['ds', 'y']

# Dublin Airport Temp
np_dat = dub_air_temp[['date', 'temp']]
np_dat.columns = ['ds', 'y']
```

#### Rain

```python
# Find the index which allows us to split off 20% of the data
cutoff_np_dar = round(np_dar.shape[0]*0.8)
cutoff_np_dar
```

```python
# Define train and test sets according to the index found above
train_np_dar = np_dar[:cutoff_np_dar]
test_np_dar = np_dar[cutoff_np_dar:]

# Print lengths
print("Train length:", len(train_np_dar))
print("Test length:", len(test_np_dar))
```

```python
# Plot of our train test split
fig, ax = plt.subplots(figsize=(20, 8))

ax.plot(train_np_dar.ds, label='Train')
ax.plot(test_np_dar.ds, label='Test')
ax.set_title('Train-Test Split');
plt.legend();
```

```python
# Initialize and fit the model on the training data
m_np_dar = NeuralProphet()

m_np_dar.fit(train_np_dar, freq='H', epochs=20)
```

```python
future_np_dar = m_np_dar.make_future_dataframe(np_dar, periods=53323)
forecast_np_dar = m_np_dar.predict(future_np_dar)
```

```python
# Extract actual values from the test set
y_true_dar = test_np_dar['y']

# Extract predicted values from the forecast
y_pred_dar = forecast_np_dar['yhat1'].iloc[:len(y_true_dar)]

# Compute MAE and RMSE
mae_npr = mean_absolute_error(y_true_dar, y_pred_dar)
rmse_npr = mean_squared_error(y_true_dar, y_pred_dar, squared=False)

# Print results
print("Neural Prophet Rain MAE: {:.4f}".format(mae_npr))
print("Neural Prophet Rain RMSE: {:.4f}".format(rmse_npr))
```
